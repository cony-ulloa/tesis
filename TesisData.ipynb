```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('participacion2017.csv', sep=';')
```

## Procesamiento de datos Participación Inventiva

```python
df['d1'] = np.where(df['d1'] == 1, 'Sí', 'No')
```

```python
df['d3'] = np.where(df['d3'] == 1, 'Sí', 'No')
```

```python
df['d5'] = np.where(df['d5'] == 1, 'Sí', 'No')
```

```python
df['d7'] = np.where((df['d7_1'] == 1) | (df['d7_2'] == 2) | (df['d7_3'] == 3), 'Sí', 'No')
```

```python
df['d7'] = np.where((df['d7_4'] != 4), 'Sí', 'No')
```

```python
df['d11'] = np.where(df['d11'] == 1, 'Sí', 'No')
```

```python
df['d13'] = np.where(df['d13'] == 1, 'Sí', 'No')
```

```python
df['d15'] = np.where(df['d15'] == 1, 'Sí', 'No')
```

```python
df['d17'] = np.where(df['d17'] == 1, 'Sí', 'No')
```

```python
df['d19'] = np.where(df['d19'] == 1, 'Sí', 'No')
```

```python
df['d21'] = np.where(df['d21'] == 1, 'Sí', 'No')
```

```python
df['d23'] = np.where(df['d23'] == 1, 'Sí', 'No')
```

```python
df['d26'] = np.where(df['d26'] == 1, 'Sí', 'No')
```

### Participacion Inventiva = Nivel Educativo

```python
df['nivel_educ'] = df['nivel_educ'].replace([1,2,3,4,5,6,7],
                    ["Básica incompleta","Básica completa", "Media Incompleta", "Media completa", "Técnica Nivel Superior", "Universitario incompleto", "Universitario completo"])
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
ax =sns.countplot( x='nivel_educ',hue='d1', data=df, palette = 'Paired')
plt.title('¿Ha sacado fotos con fines artísticos?')
ax.set(xlabel = 'Nivel Educacional', ylabel = 'Total')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, horizontalalignment = 'right')

plt.subplot(1,2,2)
ax= sns.countplot( x='nivel_educ',hue='d3', data=df, palette = 'Paired')
plt.title('¿Ha dibujado, pintado cuadros, hecho una escultura o un grabado?')
ax.set(xlabel = 'Nivel Educacional', ylabel = 'Total')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, horizontalalignment = 'right')


plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
ax=sns.countplot( x='nivel_educ',hue='d5', data=df, palette = 'Paired')
plt.title('¿Ha producido artesanías o manualidades como: tejidos, bordados, decoraciones, textiles etc..?')
ax.set(xlabel = 'Nivel Educacional', ylabel = 'Total')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, horizontalalignment = 'right')

plt.subplot(1,2,2)
ax=sns.countplot( x='nivel_educ',hue='d11', data=df, palette = 'Paired')
plt.title('¿Ha escrito cuentos, poesías o novelas?')
ax.set(xlabel = 'Nivel Educacional', ylabel = 'Total')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, horizontalalignment = 'right')

plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
ax=sns.countplot( x='nivel_educ',hue='d13', data=df, palette = 'Paired')
plt.title('¿Ha actuado, dirigido o producido alguna obra de teatro?')
ax.set(xlabel = 'Nivel Educacional', ylabel = 'Total')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, horizontalalignment = 'right')

plt.subplot(1,2,2)
ax=sns.countplot( x='nivel_educ',hue='d15', data=df, palette = 'Paired')
plt.title('¿Ha actuado, dirigido o producido alguna película, documental, cortometraje u obra audiovisual?')
ax.set(xlabel = 'Nivel Educacional', ylabel = 'Total')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, horizontalalignment = 'right')


plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
ax=sns.countplot( x='nivel_educ',hue='d17', data=df, palette = 'Paired')
plt.title('¿Ha interpretado o creado alguna obra de danza, baile o coreografía?')
ax.set(xlabel = 'Nivel Educacional', ylabel = 'Total')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, horizontalalignment = 'right')

plt.subplot(1,2,2)
ax=sns.countplot( x='nivel_educ',hue='d19', data=df, palette = 'Paired')
plt.title('¿Ha practicado malabarismo, acrobacias, clown, pasacalles, batucadas, mimo etc..?')
ax.set(xlabel = 'Nivel Educacional', ylabel = 'Total')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, horizontalalignment = 'right')


plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
ax=sns.countplot( x='nivel_educ',hue='d21', data=df, palette = 'Paired')
plt.title('¿Ha trabajado en la producción de murales o graffitis?')
ax.set(xlabel = 'Nivel Educacional', ylabel = 'Total')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, horizontalalignment = 'right')

plt.subplot(1,2,2)
ax=sns.countplot( x='nivel_educ',hue='d23', data=df, palette = 'Paired')
plt.title('¿Usted ha producido o creado contenidos para subir a Internet?')
ax.set(xlabel = 'Nivel Educacional', ylabel = 'Total')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, horizontalalignment = 'right')

plt.figure(figsize=(6,8))
plt.subplot(1,1,1)
ax=sns.countplot( x='nivel_educ',hue='d26', data=df, palette = 'Paired')
plt.title('¿Ha asistido a talleres artísticos o clases fuera del trabajo o estudio?')
ax.set(xlabel = 'Nivel Educacional', ylabel = 'Total')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, horizontalalignment = 'right')
```
